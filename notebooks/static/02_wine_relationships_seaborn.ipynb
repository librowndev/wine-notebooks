# Wine feature relationships with seaborn

This notebook focuses on static seaborn charts for a small subset of the wine features.

```python
from sklearn.datasets import load_wine
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')
wine = load_wine(as_frame=True)
df = wine.frame.copy()
df['target_name'] = df['target'].map(dict(enumerate(wine.target_names)))
features = ['alcohol', 'malic_acid', 'color_intensity', 'proline', 'target_name']
subset = df[features]
subset.head()
```

```python
sns.pairplot(subset, hue='target_name', corner=True, diag_kind='hist')
```

```python
corr = df.drop(columns=['target']).corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title('Feature correlation heatmap')
plt.tight_layout()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='target_name', y='flavanoids', hue='target_name', legend=False)
plt.title('Flavanoids by cultivar')
plt.xlabel('Cultivar')
plt.ylabel('Flavanoids')
plt.tight_layout()
```
